```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load the dataset
df=pd.read_csv('/content/netflix_titles.csv')

#displayinging info
print(df.info())
```

```python
#display first five items
df.head()
```

```python
# Data Exploration
# Number of movies and TV shows on Netflix
content_types = df['type'].value_counts()
print("Number of Movies and TV Shows on Netflix:")
print(content_types)
```

```python
# Release trends over the years
df['release_year'] = df['release_year'].fillna(0).astype(int)
release_trends = df['release_year'].value_counts().sort_index()
```

```python
# Visualization
plt.figure(figsize=(12, 6))
sns.lineplot(x=release_trends.index, y=release_trends.values, marker='o')
plt.title('Trends of Movies and TV Shows on Netflix Over the Years')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.grid(True)
plt.show()
```

```python
# Genre Analysis
# Split genres and count the frequency of each genre
genres = df['listed_in'].str.split(',').explode().str.strip()
genre_counts = genres.value_counts()
```

```python
# Visualization: Top 10 Genres
plt.figure(figsize=(12, 6))
genre_counts.head(10).plot(kind='bar', color='skyblue')
plt.title('Top 10 Genres on Netflix')
plt.xlabel('Genre')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Country-wise Analysis
# Counting the frequency of countries
countries = df['country'].str.split(',').explode().str.strip()
country_counts = countries.value_counts()
```

```python
# Visualization: Top 10 Countries
plt.figure(figsize=(12, 6))
country_counts.head(10).plot(kind='bar', color='pink')
plt.title('Top 10 Countries with Most Content on Netflix')
plt.xlabel('Country')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45, ha='right')
plt.show()
```
